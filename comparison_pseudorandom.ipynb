```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng()
samples = rng.random(size=(1024, 2))
```

```python
plt.plot(samples[:,0], samples[:,1],'.')
plt.title("1024 Pseudo-Zufallspunkte")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig('comparison_prng_plot.png', dpi=300)
plt.show()
```
