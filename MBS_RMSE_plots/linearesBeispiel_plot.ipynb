```python
x_Axis = (np.ones(10)*2) ** (np.arange(10)+6)

RMSE_results_cutted_Transposed_Log10 = np.log10(RMSE_results[:,0:].T)
x_Axis_cutted_Log10 = np.log10(x_Axis[0:])

convergence =np.log10(x_Axis**(-1/2))
convergence2 = np.log10(x_Axis**(-3/2))

plt.plot(x_Axis_cutted_Log10, RMSE_results_cutted_Transposed_Log10, 'x')

for i in range(10):
    (m, b) = np.polyfit(x_Axis_cutted_Log10, RMSE_results_cutted_Transposed_Log10[:,i:i+1], 1)
    pol = np.polyval([m, b], x_Axis_cutted_Log10)
    plt.plot(x_Axis_cutted_Log10,pol)
plt.plot(x_Axis_cutted_Log10, convergence)
plt.plot(x_Axis_cutted_Log10, convergence2)



plt.legend(['MC', 'QMC-Halton', 'S-QMC-Halton', 'QMC-Sobol', 'S-QMC-Sobol', 'MC-BB', 'QMC-Halton-BB', 'S-QMC-Halton-BB', 'QMC-Sobol-BB', 'S-QMC-Sobol-BB'][:], loc=3, prop={'size': 6})
```
