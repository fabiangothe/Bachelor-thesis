```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import qmc
```

```python
def Sobol_seq(d,m, number_paths, scramble_func):
    Sobol_seq = qmc.Sobol(d, scramble = scramble_func) # weil scramble ist noch irgendein Extraschritt und default ist True
    sample =  Sobol_seq.random_base2(m)
    return sample

n=13
samples = Sobol_seq(360,n,2**n, False)  #2**10 sind 1024
```

```python
np.shape(samples)
```

```python
plt.plot(samples[:,26], samples[:,27],'.')
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.title(str(2**n)+" Punkte")
plt.show()
```

```python
plt.plot(samples[:4096,24], samples[:4096,35],'.')
plt.title("ersten 4096 Punkte der Sobol Folge")
plt.xlabel("Dimension 25")
plt.ylabel("Dimension 36")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig('comparison_sobol_SchlechterFall1_ersten4096_plot.png', dpi=300)
plt.show()
```

```python
plt.plot(samples[4096:,24], samples[4096:,35],'.')
plt.title("4097-ter bis 8192-ter Punkt der Sobol-Folge")
plt.xlabel("Dimension 25")
plt.ylabel("Dimension 36")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig('comparison_sobol_SchlechterFall1_letzten4096_plot.png', dpi=300)
plt.show()
```

```python
plt.plot(samples[:,24], samples[:,35],'.')
plt.title("8192 Punkte der Sobol-Folge")
plt.xlabel("Dimension 25")
plt.ylabel("Dimension 36")
plt.xlim([0, 1])
plt.ylim([0, 1])
plt.savefig('comparison_sobol_SchlechterFall1_gesamt_plot.png', dpi=300)
plt.show()
```

```python
for x in np.arange(20,50,1):
    for y in np.arange(20,50,1):
        plt.plot(samples[4096:,x], samples[4096:,y],'.')
        plt.title(str(2**n)+" Punkte")
        print(x+1,y+1)
        plt.show()
```
